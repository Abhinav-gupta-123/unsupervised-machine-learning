```python
iris=datasets.load_iris()
```

```python
iris_dataset=pd.DataFrame(iris.data)
```

```python
iris_dataset.columns=iris.feature_names
```

this is the important step for all cases where we are going to use hierarichal clustering is first we should do feature scalling more better clustering

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(iris_dataset)
```

```python
X_scaled
```

```python
X_scaled.shape
```

Now we will apply pca for better clustering and keep the top 2 feature

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

so, if will write the python code to make the dindogram it will be too complex so we will use the library called scipy

```python
# %pip install scipy
```

```python
import scipy.cluster.hierarchy as sc
```

#Agglomerative clustering
#to construct a dindogram to find the number of cluster

```python
#plot the dindogram
plt.figure(figsize=(20,7))
plt.title("Dindogram")


sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title("dendogram")
plt.xlabel('sample index')
plt.ylabel('eucledian distance')
```

In above dendogram we can se there can be 2 cluster

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric="euclidean",linkage='ward')
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

Lets validate the clsuters

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=AgglomerativeClustering(n_clusters=k,metric="euclidean",linkage="ward")
    kmeans.fit(pca_scaled)
    score=silhouette_score(pca_scaled,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("number of clusters")
plt.ylabel("silhouette_coefficients")
plt.show()
```
